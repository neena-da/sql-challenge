```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt

# Import the Postgres username and password
from Keys import username, password
```

```python
# Import sqlalchemy and create engine to establish connection between the database 'EmployeeSQL' and notebook
from sqlalchemy import create_engine
engine = create_engine(f"postgresql://{username}:{password}@localhost:5432/EmployeeSQL")
connection = engine.connect()
```

```python
# Read the employees table from the database and store in a dataframe
employee_data = pd.read_sql("SELECT * FROM employees", connection)
employee_data.head(10)
```

```python
# Read the salaries table from the database and store in a dataframe
salary_data = pd.read_sql("SELECT * FROM salaries", connection)
salary_data.head(10)
```

```python
# Read the titles table from the database and store in a dataframe
title_data = pd.read_sql("SELECT * FROM titles", connection)
title_data
```

```python
# Print maximum and minimum salaries from the salaries table dataframe
print(f"Min Salary : {salary_data['salary'].min()}")
print(f"Max Salary : {salary_data['salary'].max()}")
```

```python
# Plotting histogram of Salary Ranges for Employees

# Creating bins to group Salaries
bins_size = [40000, 50000, 60000, 70000, 80000, 90000, 100000, 110000, 120000, 130000]

plt.figure(figsize=(8,8))
plt.hist(salary_data['salary'], bins=bins_size, color = 'red', alpha= 0.8, edgecolor = 'black')
plt.grid(axis='y', alpha = 0.5)
plt.xlabel('Salary Range')
plt.ylabel('Count of Employees')
plt.title('Salary ranges for employees')
plt.savefig('Output/Salaryforemployees.png')
plt.show();
```

```python
# Merging the employees dataframe with the salaries dataframe on emp_no
merge_temp_df = pd.merge(employee_data, salary_data, on='emp_no')
merge_temp_df
```

```python
# Merging the above merged dataframe with the titles dataframe on title_id
merge_df = pd.merge(merge_temp_df, title_data, left_on='emp_title_id', right_on='title_id', how = 'left')
merge_df
```

```python
# Grouping the merged dataframe on title and calculating the average salary
merge_groupby_df = merge_df.groupby(['title'])
merge_groupby_df = merge_groupby_df['salary'].mean().reset_index()
merge_groupby_df
```

```python
# Plotting a Bar Graph for the average salary per job title
plt.figure(figsize=(10, 11))
plt.bar(merge_groupby_df['title'], merge_groupby_df['salary'], color = 'red', alpha = 0.8, edgecolor = 'black')
plt.xticks(rotation = 'vertical', fontsize = 8)
plt.xlabel('Job Title')
plt.ylabel('Arverage Salary')
plt.title('Average Salary per Job Title')
plt.savefig('Output/AvgSalaryTitle.png')
plt.tight_layout()
plt.show();
```

### Observations:
* The most common Salary Range for employees is between 40000 and 49000
* We can see inconsistency in the salary per job title where the senior staff and staff have a higher average salary than managers. This would make one question the authenticity of the data in the dataset
